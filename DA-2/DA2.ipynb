```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

print("Harish P")
print("21BBS0204")
```

```python
df = pd.read_csv('/content/Naive Bayes.csv')
df.head(2)
```

```python
print("Harish P")
print("21BBS0204")

le = LabelEncoder()
categorical_cols = ['Gender', 'Existing Diet', 'Exercise', 'Strictness']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop(['Patient ID', 'Strictness'], axis=1)
y = df['Strictness']
```

```python
scaler = StandardScaler()
numerical_cols = ['Age', 'Weight', 'Height', 'BMI']
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
print("Harish P")
print("21BBS0204")

y_pred = nb_model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print("Harish P")
print("21BBS0204")
print()

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from six import StringIO
from IPython.display import Image
import pydotplus
from sklearn.tree import export_graphviz
```

```python
df = pd.read_csv('/content/Road.csv')
```

```python
x = df[['Length', 'Number_of_Bends', 'Traffic_Volume']]
y = df['Accident_Risk']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
print("Confusion Matrix:")
pd.crosstab(y_test,y_pred)
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
dot_data =StringIO()
export_graphviz(clf,out_file=dot_data,feature_names= x.columns,
                filled=True,rounded= True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
