```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/KNN.csv')
df = df.drop(columns=["Student_ID"])
df.head(2)
```

```python
df.info()
```

```python
le = LabelEncoder()
df['Drop'] = le.fit_transform(df['Drop'])
```

```python
x = df.drop('Drop',axis=1)
y = df['Drop']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.31, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
ypred = knn.predict(x_test)
ypred
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
pd.crosstab(y_test,ypred)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv('/content/Linear Reg.csv')
df.head(2)
```

```python
df['SNo'] = df.index+1
df.head(2)
```

```python
x = df['SNo']
x = x.values.reshape(-1, 1)
y = df['DEMAT_Count']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
def calculate_index(year, month):
    reference_year = 2019
    reference_month = 10
    year_diff = year - reference_year
    month_diff = month - reference_month
    total_months = year_diff * 12 + month_diff
    index = total_months + 1
    return index
result = calculate_index(2025, 1)
print(f"The index for 2025-1 is: {result}")
```

```python
model.predict([[result]])
```

```python
print(metrics.r2_score(y_test,y_pred))
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/Random For.csv')
df.head(2)
```

```python
unique_counts = df.nunique()
print(unique_counts)
```

```python
label_encoder = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = label_encoder.fit_transform(df[column])
print(df.head(2))
```

```python
x=df.drop(columns=['Fruit'])
x.head(2)
```

```python
y=df.Fruit
y.head(2)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
x_test.head(5)
```

```python
model =RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred1 = model.predict(x_test)
```

```python
print('Testing Accuracy = ', accuracy_score(y_test,y_pred1))
```

```python
pd.crosstab(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```
