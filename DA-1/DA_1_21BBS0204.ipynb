```python
print("Harish P")
print("21BBS0204")
```

```python
# 2. Heat Map
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
pp = df.pivot_table(values='Petrol_Price', index='Year', columns='Month', aggfunc='mean')
dp = df.pivot_table(values='Diesel_Price', index='Year', columns='Month', aggfunc='mean')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 15))
sns.heatmap(pp, ax=ax1, cmap='YlOrRd', annot=True, fmt='.2f')
ax1.set_title('Average Petrol Prices by Month and Year')
sns.heatmap(dp, ax=ax2, cmap='YlOrRd', annot=True, fmt='.2f')
ax2.set_title('Average Diesel Prices by Month and Year')
plt.tight_layout()
plt.show()
```

```python
print("Harish P")
print("21BBS0204")
!pip install apyori
from apyori import apriori
df = pd.read_csv(r'/content/apr1.csv')
df.head(2)
```

```python
print("Harish P")
print("21BBS0204")
print()

df = df.astype(bool)
transactions_list = df.apply(lambda x: x.index[x].tolist(), axis=1).tolist()
ar = apriori(transactions_list, min_support=0.4, min_confidence=0.7)
results = list(ar)
print("Number of association rules:", len(results))
max_elements = max(len(result.items) for result in results)
print(f"\nRules with the maximum number of elements {max_elements}:")
print("Final Answer")
for i in results:
  if len(i.items) == max_elements:
    print(list(i.items))
print()

print("All rules following the Conditions")
def inspect(results, max_elements):
    for result in results:
        items = [x for x in result.items]
        if len(items) == max_elements:
            print(f"Rule: {items}")
            print(f"Support: {result.support:.4f}")
            for rule in result.ordered_statistics:
                print(f"Confidence: {rule.confidence:.4f}")
                print(f"Lift: {rule.lift:.4f}")
                print(f"{list(rule.items_base)} -> {list(rule.items_add)}")
            print('\n')

inspect(results, max_elements)
```
